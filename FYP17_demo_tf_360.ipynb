# Robotic angle model training #yangerz way
INPUTS: First image showing the object unhindered by the robot grasper + angle it grasped at <br>
OUTPUTS: The predicted success/fail for grasping that image.

```python
import tensorflow as tf
import numpy as np
import os
import pandas as pd
from tqdm import tqdm
WIDTH=80
IMG_SIZE=WIDTH
HEIGHT=80
EPOCHS=100
LR=1e-3
directory='./model_success_prediction/'
MODEL_NAME='FYP17_360_yangerz-{}{}.model'.format(LR,'_V2')
train_data = np.load('training_data_all_360.npy')
print('done')
```

## Defining the Model
A simple 4 layer CNN with 2 FCL is used for model training/prediction. 
The amount of data isn't complex so we won't need a complex model.

```python
import tflearn
from tflearn.layers.conv import conv_2d, max_pool_2d
from tflearn.layers.core import input_data, dropout, fully_connected
from tflearn.layers.estimator import regression
from tflearn.layers.normalization import batch_normalization

import tensorflow as tf
tf.reset_default_graph()

convnet = input_data(shape=[None, IMG_SIZE, IMG_SIZE, 1], name='input1') 

convnet = conv_2d(convnet, 64, 5, activation='relu')
#convnet = max_pool_2d(convnet, 5)
convnet = batch_normalization (convnet)

convnet = conv_2d(convnet, 128, 5, activation='relu')
#convnet = max_pool_2d(convnet, 5)
convnet = batch_normalization (convnet)
convnet = dropout(convnet, 0.7)

convnet = conv_2d(convnet, 256, 5, activation='relu')
#convnet = max_pool_2d(convnet, 5)
convnet = batch_normalization (convnet)

convnet = conv_2d(convnet, 128, 5, activation='relu')
convnet = max_pool_2d(convnet, 5)
convnet = batch_normalization (convnet)
convnet = dropout(convnet, 0.7)

################SECOND BRANCH FOR ANGLE#####

anglenet=input_data(shape=[None,1,1,1], name='input2') 

anglenet= fully_connected(anglenet, 128, activation='relu')
anglenet = dropout(anglenet, 0.8)


################merged branches
mergenet= tflearn.layers.merge_ops.merge([tf.tile(anglenet,[1,1]),convnet], 'elemwise_sum', axis=1, name='Merge')

mergenet = conv_2d(mergenet, 64, 5, activation='relu')
#mergenet = max_pool_2d(mergenet, 5)
mergenet = batch_normalization (mergenet)
mergenet = dropout(mergenet, 0.7)

mergenet = conv_2d(mergenet, 64, 5, activation='relu')
mergenet = max_pool_2d(mergenet, 5)
mergenet = batch_normalization (mergenet)
mergenet = dropout(mergenet, 0.7)

mergenet= fully_connected(mergenet, 64, activation='relu')
mergenet = dropout(mergenet, 0.8)

mergenet= fully_connected(mergenet, 64, activation='relu')
mergenet = dropout(mergenet, 0.8)

mergenet = fully_connected(mergenet, 2, activation='softmax')
mergenet = regression(mergenet, optimizer='adam', learning_rate=LR, loss='categorical_crossentropy', name='targets')

model = tflearn.DNN(mergenet, tensorboard_dir='C:\Tensorboard_logs\FYP17_yangerz\log')
```

## Begin Training!
We use TFlearn because it's easily implementable and requires little effort. The model fits over a set number of epochs and the model is saved every EPOCH

We use 15% of the model for training and 85% for training to prevent overfitting.

```python
directory='./model_success_prediction/'
for i in range(EPOCHS):
    train_data = np.load('training_data_all_360.npy')

    train = train_data[:-int(len(train_data)*0.15)]
    test = train_data[-int(len(train_data)*0.15):]

    X1 = np.array([i[0] for i in train]).reshape(-1,WIDTH,HEIGHT,1)#image
    X2 = np.array([i[2] for i in train]).reshape(-1,1) #angle scalar
    Y = [i[3] for i in train]

    test_x1 = np.array([i[0] for i in test]).reshape(-1,WIDTH,HEIGHT,1)
    test_x2= np.array([i[2] for i in test]).reshape(-1,1)
    test_y = [i[3] for i in test]

    model.fit({'input1': X1,'input2': X2}, {'targets': Y}, n_epoch=1, validation_set=({'input1': test_x1,'input2': test_x2}, {'targets': test_y}), 
        snapshot_step=10000, show_metric=True, run_id=MODEL_NAME)

    #tensorboard --logdir=foo:C:/Tensorboard_logs/FYP17/log

    model.save(directory + MODEL_NAME)
```

## Testing for Accuracy
Accuracy is simply number of correct predictions within the test set divided by the total number of test inputs

```python
accuracy_sum=0
n=0
for num,data in enumerate(test):
    n=n+1
    
    Y_actual=np.argmax(data[3])    
    img_data = data[0]
    X2=data[2] #Angle
    X2=np.asarray(X2).reshape(1,1)
    X1 = img_data.reshape(1,IMG_SIZE,IMG_SIZE,1)
    model_out = model.predict(X={'input1': X1,'input2': X2})[0]
    Y_predicted=np.argmax(model_out)
    if Y_actual==Y_predicted:       
        accuracy_sum=accuracy_sum+ 1 #define accuracy as (1-delta)/360
accuracy=accuracy_sum/n
accuracy
```
